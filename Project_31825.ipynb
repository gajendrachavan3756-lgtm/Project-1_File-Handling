```python
import os
import json
import csv
```

```python
def load_data(main_folder):
    sales_data = {}
    product_details = {}
    product_descriptions = {}

    sales_file = os.path.join(main_folder, "sales_data.csv")
    with open(sales_file, 'r', newline='') as f:
      reader = csv.reader(f)
      next(reader)
      for index, row in enumerate(reader, start=1):
          sku = row[0]
          sales = list(map(int, row[1:15]))
          sales_data[sku] = sales


    details_folder = os.path.join(main_folder, "product_details")
    if os.path.exists(details_folder):
        for filename in os.listdir(details_folder):
            if filename.endswith(".json"):
                filepath = os.path.join(details_folder, filename)
                with open(filepath) as f:
                    sku = filename.replace("details_", "").replace(".json", "")
                    product_details[sku] = json.load(f)

    description_folder = os.path.join(main_folder, "product_descriptions")
    if os.path.exists(description_folder):
        for filename in os.listdir(description_folder):
            if filename.endswith(".txt"):
                filepath = os.path.join(description_folder, filename)
                with open(filepath) as f:
                    sku = filename.replace("details_", "").replace(".txt", "")
                    product_descriptions[sku] = f.read().strip()

    return sales_data, product_details, product_descriptions

```

```python
def update_sales_data(sales_data, sku, sales_list):
    sales_data[sku] = sales_list

def update_product_details(product_details, sku, Product_name, brand, model, specifications, price, availability):
    product_details[sku] = {
        "Product_name": Product_name,
        "brand": brand,
        "model": model,
        "specifications": specifications,
        "price": price,
        "availability": availability
    }

def update_product_description(product_descriptions, sku, description):
    product_descriptions[sku] = description
```

```python
def dump_data(sales_data, product_details, product_descriptions, main_folder):
    os.makedirs(main_folder, exist_ok=True)

    # Dump sales_data.csv
    sales_file = os.path.join(main_folder, "sales_data.csv")
    with open(sales_file, mode='w', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(["product_SKU", "Day1","Day2", "Day3", "Day4", "Day5", "Day6","Day7", "Day8", "Day9", "Day10","Day11" ,"Day12", "Day13","Day14"])
        for sku, sales in sales_data.items():
            writer.writerow([sku] + sales)

    # Dump product_details JSON files
    details_folder = os.path.join(main_folder, "product_details")
    os.makedirs(details_folder, exist_ok=True)
    for sku, details in product_details.items():
        filepath = os.path.join(details_folder, f"details_{sku}.json")
        with open(filepath, 'w') as f:
            json.dump(details, f, indent=2)

    # Dump product_descriptions TXT files
    description_folder = os.path.join(main_folder, "product_discriptions")
    os.makedirs(description_folder, exist_ok=True)
    for sku, description in product_descriptions.items():
        filepath = os.path.join(description_folder, f"details_description{sku}.txt")
        with open(filepath, 'w') as f:
            f.write(description)
```
